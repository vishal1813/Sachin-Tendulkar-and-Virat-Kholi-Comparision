## Both Combined

```python
dfz=df1.append(df2)

dfz
```

```python
dfz['Innings Runs Scored Num'].unique()
```

```python
#clearing out 'nan' and'-'
dfz=dfz.dropna(subset=['Innings Runs Scored Num'])

dfz['Innings Runs Scored Num'].unique()
```

```python
dfz=dfz[dfz['Innings Runs Scored Num']!='-']

dfz['Innings Runs Scored Num'].unique()
```

```python
dfz['Innings Date']=pd.to_datetime(dfz['Innings Date'])

dfz['year']=dfz['Innings Date'].dt.year

dfz.dtypes
```

```python
#converting to integer type
dfz['Innings Runs Scored Num']=dfz['Innings Runs Scored Num'].astype('int')
dfz['Innings Balls Faced']=dfz['Innings Balls Faced'].astype('int')

sachin_df=dfz[(dfz.year>=1989)&(dfz.year<=2012)]
virat_df=dfz[(dfz.year>=2008)&(dfz.year<=2019)]
```

## All of Sachin's innings

```python
sdf=dfz[dfz['Innings Player']=='SR Tendulkar']

sdf
```

## All of Virat's innings

```python
vdf=dfz[dfz['Innings Player']=='V Kohli']

vdf
```

```python
#average
sachin_avg=sum(sdf['Innings Runs Scored Num'])/len(sdf)
virat_avg=sum(vdf['Innings Runs Scored Num'])/len(vdf)

[sachin_avg,virat_avg]
```

```python
#strikerates
ssr=100*sum(sdf['Innings Runs Scored Num'])/sum(sdf['Innings Balls Faced'])
vsr=100*sum(vdf['Innings Runs Scored Num'])/sum(vdf['Innings Balls Faced'])

[ssr,vsr]
```

```python
#team contribution
scontribu=100*sum(sdf['Innings Runs Scored Num'])/sum(sachin_df[sachin_df['Country']=='India']['Innings Runs Scored Num'])
vcontribu=100*sum(vdf['Innings Runs Scored Num'])/sum(virat_df[virat_df['Country']=='India']['Innings Runs Scored Num'])


[scontribu,vcontribu]
```

## Sachin's Competitors

## Virat's Competitors

```python
dfz=dfz.dropna(subset=['100s'])
dfz=dfz.dropna(subset=['50s'])

sdf=sdf.dropna(subset=['100s'])
vdf=vdf.dropna(subset=['100s'])
sdf=sdf.dropna(subset=['50s'])
vdf=vdf.dropna(subset=['50s'])


nsachin_df=dfz[dfz['Innings Player']!='SR Tendulkar']
nvirat_df=dfz[dfz['Innings Player']!='V Kholi']
```

```python
#comapring sachin and virat's average with average of players of their era
nsavg=sum(nsachin_df['Innings Runs Scored Num'])/len(nsachin_df)
nvavg=sum(nvirat_df['Innings Runs Scored Num'])/len(nvirat_df)

sachin_avg/nsavg,virat_avg/nvavg
```

```python
#comapring sachin and virat's strike rate with average strike rate of their era
nssr=100*sum(nsachin_df['Innings Runs Scored Num'])/sum(nsachin_df['Innings Balls Faced'])
nvsr=100*sum(nvirat_df['Innings Runs Scored Num'])/sum(nvirat_df['Innings Balls Faced'])

ssr/nssr,vsr/nvsr
```

```python
v50s=sum(vdf['50s'])
s50s=sum(sdf['50s'])
s50s,v50s
```

```python
#comparing 50s scored during their eras
nv50s=sum(nvirat_df['50s'])
ns50s=sum(nsachin_df['50s'])
100*s50s/ns50s,100*v50s/nv50s
```

```python
#comparing 100s scored during their eras
v100s=sum(vdf['100s'])
s100s=sum(sdf['100s'])
nv100s=sum(nvirat_df['100s'])
ns100s=sum(nsachin_df['100s'])
100*s100s/ns100s,100*v100s/nv100s
```

```python
finalcompar=pd.DataFrame({'Parameter':['Team Contribution','50s','100s','StrikeRate','Average'],
                          'Sachin R Tendulkar':[scontribu,100*s50s/ns50s,100*s100s/ns100s,ssr/nssr,sachin_avg/nsavg],
                          'Virat Kholi':[vcontribu,100*v50s/nv50s,100*v100s/nv100s,vsr/nvsr,virat_avg/nvavg],'WINNER':['Virat Kholi','Sachin R Tendulkar','Sachin R Tendulkar','Virat Kholi','Virat Kholi']})
```
